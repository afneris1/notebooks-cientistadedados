```python
#Importando Biblioteca
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as sts
```

```python
#Criando DataFrame
dataset = pd.read_csv('./dados.csv',sep=';')
```

```python
#Removendo Valores Duplicados
dataset.drop_duplicates(['MUNICIPIO'],inplace=True)
```

```python
#Boxplot PIB
dataset.loc[dataset['PIB'] > 100000, 'PIB'] = sts.median(dataset['PIB'])
sns.boxplot(data=dataset['PIB']).set_title('PIB');
```

```python
#Boxplot Valor de Empenho
dataset.loc[dataset['VALOREMPENHO'] > 1300000, 'VALOREMPENHO'] = sts.median(dataset['VALOREMPENHO'])
sns.boxplot(data=dataset['VALOREMPENHO']).set_title('VALOR EMPENHO');
```

```python
#Histograma PIB
sns.histplot(data=dataset, x='VALOREMPENHO', kde=True).set_title('PIB');
```

```python
#Histograma Valor de Empenho
sns.histplot(data=dataset, x='VALOREMPENHO', kde=True).set_title('VALOR EMPENHO');
```

```python
#Top 10  Maiores PIB
ax = sns.barplot(data=dataset.sort_values('PIB').tail(10),x='MUNICIPIO',y='PIB',palette ='gray')
ax.figure.set_size_inches(12, 8)
ax.tick_params(axis='x', rotation=90)
ax.set_title('Top 10  Maiores PIB')
ax.set_ylabel('PIB')
ax.set_xlabel('Município');
```

```python
#Top 10  Menores PIB
ax = sns.barplot(data=dataset.sort_values('PIB').head(10),x='MUNICIPIO',y='PIB',palette ='gray')
ax.figure.set_size_inches(12, 8)
ax.tick_params(axis='x', rotation=90)
ax.set_title('Top 10  Menores PIB')
ax.set_ylabel('PIB')
ax.set_xlabel('Município');
```

```python
#Top 10  Menores Valores de Empenho
ax = sns.barplot(data=dataset.sort_values('VALOREMPENHO').head(10),x='MUNICIPIO',y='VALOREMPENHO',palette ='gray')
ax.figure.set_size_inches(12, 8)
ax.tick_params(axis='x', rotation=90)
ax.set_title('Top 10  Menores Valores de Empenho')
ax.set_ylabel('Valor de Empenho')
ax.set_xlabel('Município');
```

```python
#Top 10  Maiores Valores de Empenho
ax = sns.barplot(data=dataset.sort_values('VALOREMPENHO').tail(10),x='MUNICIPIO',y='VALOREMPENHO',palette ='gray')
ax.figure.set_size_inches(12, 8)
ax.tick_params(axis='x', rotation=90)
ax.set_title('Top 10  Maiores Valores de Empenho')
ax.set_ylabel('Valor de Empenho')
ax.set_xlabel('Município');
```
